### Task 2: Preprocessing Function

```python
import re

MEANINGFUL_SHORT = {"no", "not"}

def preprocess_text(text):
    
    # Error Handling
    if not isinstance(text, str) or text.strip() == "":
        return [], ""
    
    # Remove URLs and emails
    text = re.sub(r"http\S+|www\S+|\S+@\S+", "", text)
    
    # Remove numbers
    text = re.sub(r"\d+", "", text)
    
    # Convert to lowercase
    text = text.lower()
    
    # Handle repeated characters
    text = re.sub(r"(.)\1{2,}", r"\1\1", text)
    
    # Remove special characters and emojis
    text = re.sub(r"[^\w\s]", "", text)
    
    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()
    
    # Tokenization
    tokens = text.split()
    
    # Remove short tokens except "no", "not"
    tokens = [word for word in tokens if len(word) > 2 or word in MEANINGFUL_SHORT]
    
    clean_sentence = " ".join(tokens)
    
    return tokens, clean_sentence
```

### Task 3: Stress Testing

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for s in sentences:
    tokens, clean = preprocess_text(s)
    print("\nOriginal:", s)
    print("Tokens:", tokens)
    print("Cleaned:", clean)
```

### Task 4: Token Analytics

```python
def token_analysis(tokens):
    if len(tokens) == 0:
        return 0, 0, 0
    
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = round(sum(len(t) for t in tokens) / total, 2)
    
    return total, unique, avg_len


results = []

for s in sentences:
    tokens, _ = preprocess_text(s)
    total, unique, avg = token_analysis(tokens)
    
    results.append((s, total))
    
    print("\nSentence:", s)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", avg)


# Analysis
most_noise = min(results, key=lambda x: x[1])
most_meaningful = max(results, key=lambda x: x[1])

print("\nMost noisy sentence:", most_noise[0])
print("Most meaningful sentence:", most_meaningful[0])
```

### Task 5: Frequency Analysis


```python
from collections import Counter

all_tokens = []

for s in sentences:
    tokens, _ = preprocess_text(s)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("\nTop 10 Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

### Task 6: Full Pipeline

```python
def full_pipeline(text_list):
    
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


output = full_pipeline(sentences)
print(output)
```
